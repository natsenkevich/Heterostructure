```python
d_MoS2 = 6.21
Eb=np.zeros(5)

#for n in (1, n_MoS2 + 1, 1):
for n in [1, 2, 3, 4, 5]:
    
    d = [6.21 for i in range(n - 1)]
    
    hl_array = np.zeros(2 * n)  # the factor 2 accounts for
    el_array = np.zeros(2 * n)  # the dipole component
    hl_array[0] = 1  # the hole is placed on MoS2
    el_array[0] = 1  # the electron is  placed on MoS2
    
    HS = Heterostructure(structure=['%dH-MoS2' % n],  # set up structure
                         d=d,                         # layer distance array
                         include_dipole=True,
                         wmax=0,                      # only include w=0
                         d0=6.21)                     # width of single layer
    ee, ev = HS.get_exciton_binding_energies(eff_mass=0.277,  # exciton eff_mass
                                             e_distr=el_array,
                                             h_distr=hl_array)
    
    Eb[n-1]=-ee[0].real
    
n = np.arange(1, 6, 1)
plt.plot(n, Eb, '-s')
plt.title('Exciton Binding Energy', fontsize=20)
plt.xlabel(r'no. MoS2', fontsize=20)
plt.ylabel(r'$E_{\rm b}$ (eV)', fontsize=20)
plt.savefig('MoS2_on_hBN_Eb.svg')
plt.show()
```
